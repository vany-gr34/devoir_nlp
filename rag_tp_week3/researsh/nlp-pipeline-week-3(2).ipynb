## 3. Extract texts : our knowledge base for RAG

Transformation de format des données car FAISS + embeddings attend une LISTE de textes et non pas un DataFrame

```python
texts = df["infraction_desc"].tolist()
```

## 4. Embedding model : the semantic understanding layer

This model transforms text into vectors (numbers).

Example:

"STOP sign" → [0.12, -0.44, 0.88, ...]


**Why ? LLMs cannot search text directly, so we compare meaning using vectors**

```python
from sentence_transformers import SentenceTransformer
import numpy as np

embedding_model = SentenceTransformer("sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2")
# explore more embedding models on : https://huggingface.co/sentence-transformers/all-MiniLM-L6-v2
```

## 5. Build FAISS index

- Step 1: Encode all texts => Each rule becomes a vector.
- Step 2: Create FAISS index => A structure optimized for similarity search.
- Step 3: Store vectors => FAISS now holds all knowledge.

----
What FAISS does conceptually:

- Instead of: “search words”
- I does : “search meaning”

```python
import faiss

embeddings = embedding_model.encode(texts)

dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(np.array(embeddings))

print("FAISS index size:", index.ntotal)
```

## 6. Retrieval function

```python
def retrieve(query, k=3):
    # Convert query to vector
    query_vec = embedding_model.encode([query])
    # Search similar documents
    distances, indices = index.search(np.array(query_vec), k)
    # FAISS returns: closest vectors, their positions
    return [texts[i] for i in indices[0]]
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

model_name = "microsoft/Phi-4-mini-instruct"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto")

# Pipeline setup
# This wraps the model into a simple function: input → prompt, output → text
generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer
)
```

```python
def rag_answer(query):
    # Retrieve context
    docs = retrieve(query)
    # Build context : Combine retrieved documents into one block
    context = "\n".join(docs)
    # Build prompt
    # We give the LLM: background knowledge, user question
    prompt = f"""
Role: Moroccan traffic law expert.

Instructions:
- Answer strictly from the context.
- Do not add external knowledge.
- If missing info, say: "لا أملك معلومات كافية في السياق."
- Answer in Arabic only.
- Keep answers short and precise.

Context:
{context}

User Question:
{query}

Final Answer:
"""
    # Generate answer
    output = generator(
        prompt,
        max_new_tokens=150,
        do_sample=True,
        temperature=0.3
    )

    return output[0]["generated_text"]
```

### 9. Chat interface

```python
def chat_rag():
    print("🚗 RAG Code de la Route (tape 'exit' pour quitter)\n")
    
    while True:
        query = input(" Question: ")
        
        if query.lower() == "exit":
            print("Fin du chatbot")
            break
        
        answer = rag_answer(query)
        
        print("\nRéponse:\n", answer)
        print("\n" + "-"*50 + "\n")

chat_rag()
```

## 10. Comparaison des LLMs — Évaluation qualitative manuelle

```python
# ── Questions de test ──────────────────────────────────────────
# Ces questions couvrent différents aspects du code de la route marocain

test_questions = [
    "ما هي عقوبة تجاوز الإشارة الحمراء؟",
    "ما هو الحد الأقصى للسرعة داخل المدن؟",
    "ما هي عقوبة قيادة السيارة بدون رخصة؟",
    "ما هي عقوبة استخدام الهاتف أثناء القيادة؟",
    "ماذا تعني رخصة دولية للسياقة؟",
]

print(f"✅ {len(test_questions)} questions de test prêtes.")
```

```python
# ── Génération des réponses avec Phi-4-mini (déjà chargé) ─────
# Pour tester un autre modèle, recharge generator avec un nouveau model_name
# et relance cette cellule en changeant le label dans results

import time
import pandas as pd

CURRENT_MODEL_LABEL = "Phi-4-mini"   # ← change ici si tu testes un autre modèle

results = []

for q in test_questions:
    start = time.time()
    raw = rag_answer(q)
    elapsed = round(time.time() - start, 2)
    
    # Extrait uniquement la partie après "Final Answer:"
    answer = raw.split("Final Answer:")[-1].strip()
    
    results.append({
        "Modèle"        : CURRENT_MODEL_LABEL,
        "Question"      : q,
        "Réponse"       : answer,
        "Temps (s)"     : elapsed,
        "Pertinence /5" : "",   # à remplir dans la cellule suivante
        "Cohérence /5"  : "",
        "Exactitude /5" : "",
    })
    print(f"Q: {q}")
    print(f"→ {answer[:120]}...")
    print(f"   ⏱ {elapsed}s\n")

print(f"✅ {len(results)} réponses générées pour [{CURRENT_MODEL_LABEL}].")
```

```python
# ── Scores manuels ─────────────────────────────────────────────
# Lis les réponses ci-dessus et note chaque réponse sur 5 :
#   Pertinence : la réponse répond-elle bien à la question ?
#   Cohérence  : l'arabe est-il bien formulé ?
#   Exactitude : l'information correspond-elle au CSV ?

# Format : (index_question, pertinence, cohérence, exactitude)
scores = [
    (0,  4, 4, 4),
    (1,  5, 4, 5),
    (2,  3, 4, 3),
    (3,  4, 5, 4),
    (4,  4, 4, 4),
]

for idx, p, c, e in scores:
    results[idx]["Pertinence /5"] = p
    results[idx]["Cohérence /5"]  = c
    results[idx]["Exactitude /5"] = e

print("✅ Scores saisis.")
```

```python
# ── Tableau de comparaison final ───────────────────────────────
df = pd.DataFrame(results)

# Calcul du score moyen
df["Score moyen /5"] = df[["Pertinence /5", "Cohérence /5", "Exactitude /5"]].mean(axis=1)

# ── Tableau détaillé
print("=== Réponses détaillées ===")
display(df[["Modèle", "Question", "Réponse", "Pertinence /5", "Cohérence /5", "Exactitude /5", "Temps (s)"]])

# ── Résumé par modèle
summary = df.groupby("Modèle")[["Pertinence /5", "Cohérence /5", "Exactitude /5", "Score moyen /5", "Temps (s)"]].mean().round(2)

print("\n=== Résumé par modèle ===")
display(summary.style
    .background_gradient(cmap="YlGn", axis=0)
    .format("{:.2f}")
    .set_caption("Comparaison qualitative des LLMs — scores /5")
)
```

```python
# ── Export CSV (optionnel) ─────────────────────────────────────
df.to_csv("comparison_llm_results.csv", index=False, encoding="utf-8-sig")
print("✅ Sauvegardé : comparison_llm_results.csv")
```
